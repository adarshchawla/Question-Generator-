```python
# type 2

import random

i=0
#while i<100:

names1 = ['Karen', 'Susan', 'Amanda', 'Mary']
name1 = random.choice(names1)

names2 = ['Billie', 'Taylor', 'Pamela', 'April']
name2 = random.choice(names2)

ratio1 = list(range(2,6))
random.shuffle(ratio1)
ratio_1= random.choice(ratio1)

ratio2 = list(range(6,10))
random.shuffle(ratio2)
ratio_2 = random.choice(ratio2)

agegap = list(range(4,15))
random.shuffle(agegap)
age_gap = random.choice(agegap) 


futureage = list(range (18,30))
random.shuffle(futureage)
future_age = random.choice(futureage)



print('At present, the ratio between the ages of' ,name1, 'and' , name2 , 'is', str(ratio_1),' : ', str(ratio_2))
print('After', str(age_gap) , 'years,', name1, 'age will be', str(future_age), ' years. What is the age of',name2, 'at present ?')

#solution

x = ((future_age- age_gap)/ratio_1)


value1 = x*ratio_1
value2 = x*ratio_2

print('Solution')
print("Their ages are" , str(value1),'and', str(value2), 'respectively')

#i=i+1
```
